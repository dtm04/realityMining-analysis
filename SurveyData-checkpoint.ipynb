```python
df1 = pd.DataFrame(mat['s']['neighborhood'])
df1.dtypes
```

```python
df1 = pd.DataFrame(mat['s']['neighborhood'])
df2 = pd.DataFrame(mat['s']['my_hangouts'])
df3 = pd.DataFrame(mat['s']['my_travel'])
df4 = pd.DataFrame(mat['s']['my_hours'])
frames = [df1, df2, df3, df4]
df = pd.concat(frames)
df = df.transpose()
df.columns = ['neighborhood', 'my_hangouts', 'my_travel', 'my_hours']

# make the data more readable
for col in df:
    df[col] = df[col].astype(str)
    df[col] = df[col].map(lambda x: str(x).lstrip('[').rstrip(']'))
    df[col] = df[col].map(lambda x: re.findall(r"\['?([^'\]]+)'?\]", x))
    df[col] = df[col].map(lambda x: str(x).lstrip('[').rstrip(']'))

df.replace('', np.nan, inplace=True) 
df
```

```python
df['neighborhood'].value_counts().plot(kind='bar')
savefig('areas.png', bbox_inches='tight', dpi=300, format='png')
```

```python
df['my_travel'].value_counts().plot(kind='pie')
savefig('travel.png', bbox_inches='tight', dpi=300, format='png')
```
