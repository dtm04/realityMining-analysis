```python
#%matplotlib notebook
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import community
import operator
from itertools import chain
from networkx.algorithms import bipartite
from pylab import *
```

```python
# read in the graph 
G = nx.read_edgelist("ia-reality\ia-reality.mtx", comments='%')
print (list(G.edges())[0:5])
```

```python
plt.figure(figsize=(10,9))
nx.draw_networkx(G, with_labels=False)
```

```python
# number of nodes
len(G)
```

```python
# density
nx.density(G)
# density = .0003, not unusual, most complex networks have low density
```

```python
# nodes with highest degree (most popular nodes)
dgr = nx.degree_centrality(G)
sorted(dgr, key=dgr.get, reverse=True)[:4]
# Nodes with highest degree: 65, 15, 37, 59
```

```python
#degree-centrality of node 65
print(dgr.get('65'))
```

```python
ego = '65'
friends = G[ego]
print(len(friends))
```

```python
# ego network
egonet = nx.ego_graph(G, ego)
nx.draw_networkx(egonet, with_labels=False)
```

```python
#clutering coefficient
cc = nx.clustering(G, ego)
print(cc)
```

```python
# avg clustering coefficient for the entire graph
acc = nx.average_clustering(G)
print(acc)
```

```python
# graph of the main core (k-core)
KC = G.subgraph(nx.k_core(G).nodes())
plt.figure(figsize=(10,9))
nx.draw_networkx(KC)
savefig('k-core.png', bbox_inches='tight', dpi=300, format='png')
```

```python
len(KC)
```

```python
CG = nx.make_max_clique_graph(G)
len(CG)
```

```python
# Partition in to communities using the Louvain algorithm
part = community.best_partition(G)
# Puts discovered communities into a dictionary, node labels are keys and community ID as value
```

```python
community.modularity(part, G)
# modularity > 0.6 corresponds to a visible community structure
# modularity = 0.87
```

```python
# induced subgraph replaces the communities of the graph with synthetic nodes
subgraph = community.induced_graph(part, G)
plt.figure(figsize=(10,9))
nx.draw_networkx(subgraph)
savefig('synthetic.png', bbox_inches='tight', dpi=300, format='png')
```

```python
# How large are the communities?
part_series = pd.Series(part)
part_series.value_counts()
# Community ID 20 is the largest with 307 members
```

```python
gccs_nodes = chain.from_iterable(sorted(nx.connected_components(G), key=len)[-1:])
gccs = nx.subgraph(G, gccs_nodes)
len(gccs)
# page 156
```

```python
community_df = pd.DataFrame(part_series.value_counts())
community_df.columns = ['size'] 
print (len(community_df))
community_df
```
